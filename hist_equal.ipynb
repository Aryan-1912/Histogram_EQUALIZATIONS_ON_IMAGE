```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("./car.jpg")

if image is None:
    print("issue with image loading")
else:
    print("loaded successfully")
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
hist, bins = np.histogram(image_grayscale.flatten(), 256, [0,256])


# histogram equalization 

equalized_img = cv2.equalizeHist(image_grayscale)
equalized_hist, bins = np.histogram(equalized_img.flatten(), 256, [0,256])


plt.figure(figsize=(12,8))

plt.subplot(2,2,1 )
plt.imshow(image_grayscale, cmap='gray')
plt.title("original image")

plt.subplot(2,2,2)
plt.plot(hist, color = 'blue')
plt.title("original image histogram ")

plt.subplot(2,2,3 )
plt.imshow(equalized_img, cmap='gray')
plt.title("equalized  image")

plt.subplot(2,2,4 )
plt.plot(equalized_hist, color='green')
plt.title("equalized image histogram")



plt.tight_layout()
plt.show()



```

```python
# LAB 
# Y U V
# we will hist y and then join uv at the end and can get rgb after this 
# by this we can perform equalization in rgb images 


image_yuv = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)

# [row, cols, index of channel --> 0 means targeting only y of (y,u,v)]
image_yuv[:,:,0]=cv2.equalizeHist(image_yuv[:,:,0])

# yuv to bgr
equ_col_ig = cv2.cvtColor(image_yuv, cv2.COLOR_YUV2BGR)


# bgr to rgb
equ_col_rgb = cv2.cvtColor(equ_col_ig, cv2.COLOR_BGR2RGB)


plt.figure(figsize=(12,8))

plt.subplot(1,2,1 )
plt.imshow(image_rgb)
plt.title("original image")

plt.subplot(1,2,2)
plt.imshow(equ_col_rgb)
plt.title("Equalised RGB image  ")



plt.tight_layout()
plt.show()





```
